```python
import pickle
X = pickle.load(open('/content/drive/MyDrive/Potato_X.pickle', 'rb'))
y = pickle.load(open('/content/drive/MyDrive/Potato_y.pickle', 'rb'))
```

```python
X.shape
```

```python
from keras.layers import Input, Lambda, Dense, Flatten, Dropout
from keras.models import Model
from keras.applications.resnet_v2 import ResNet152V2
from keras.applications.resnet_v2 import preprocess_input
from keras.preprocessing import image
from keras.models import Sequential
import numpy as np
```

<h3>ResNet152V2</h3>

```python
resnet = ResNet152V2(input_shape=(224,224,3), weights='imagenet', include_top=False)
```

```python
y = y.reshape(4682, 1)
```

```python
y.shape
```

```python
for layer in resnet.layers:
  layer.trainable = False
```

```python
x = Flatten()(resnet.output)
```

<h3>ANN</h3>

```python
x = Dense(128, activation='relu', name='fcl')(x)
```

```python
x = Dropout(0.3)(x)
```

```python
prediction = Dense(3, activation='softmax', name='prediction')(x)
```

```python
model = Model(inputs=resnet.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(
  loss='sparse_categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
model.fit(X, y, batch_size=16, epochs=5, validation_split=0.2)
```
